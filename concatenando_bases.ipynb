```python
# Lendo as 11 bases .xlsx com os nomes de 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023 e 2024
folder_path = 'bases_concatenar/'
file_name = ''
file_extension = '.xlsx'
dfs = []
for i in range(2014, 2025):
    file_path = folder_path + file_name + str(i) + file_extension
    df = pd.read_excel(file_path)
    dfs.append(df)
```
